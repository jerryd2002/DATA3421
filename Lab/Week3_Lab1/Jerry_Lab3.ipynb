# Question 1, Imputation Techniques Comparison:


```python
#Mean Imputation 
meanimp = df.select_dtypes(include=['number']).columns
df[meanimp] = df[meanimp].fillna(df[meanimp].mean())
print(df[meanimp].describe())

#Median Imputation 
medianimp = df.select_dtypes(include=['number']).columns
df[medianimp] = df[medianimp].fillna(df[medianimp].median())
print(df[medianimp].describe())


#Mode imputation for a categorical column
mode_value = df['Year of Release'].mode()[0]
df['Year of Release'].fillna(mode_value, inplace=True)
print(df['Year of Release'].describe())

```
